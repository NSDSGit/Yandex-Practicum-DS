## Предобработка данных

### Переименование столбцов

```python
df.rename(columns = {'cityCenters_nearest' : 'city_centers_nearest'}, inplace = True)# переименуем только один столбец, остальное вроде приемлемо
```

### Заполнение пропусков
Пойдём по порядку

```python
df['is_apartment'].unique() # смотрим из чего состоит палитра значений
```

```python
df['balcony'].unique() # столбец с похожей проблемой решил проверить тут же
```

Резонно предположить, что nan в данном случае означает, что всё-таки это не апартаменты и отсутствие балкона, поэтому заменять будем на False и 0 соответственно

```python
df['is_apartment'] = df['is_apartment'].fillna('False')
df['balcony'] = df['balcony'].fillna(0)
```

Порядок. Идём дальше

```python
df['floors_total'].unique() # в целом всё прилично, nan-ов немного, можно их и отбросить. И сразу проверим ещё одного подобного кандидата
df['locality_name'].sort_values().unique() # здесь, возможно, произошёл сбой, отсюда и nan-ы. Также есть где и в плане категоризации разгуляться... лечить будем обоих
df = df[df['floors_total'].notna()]
df = df[df['locality_name'].notna()]
```

Готово, потеряли мы совсем немного

Далее, разберёмся с площадями: 
* Пропуски в столбцах living_area и kitchen_area, не имеют какой-либо выраженной закономерности по категориям недвижимости, но резонно предположить, что в квартирах-студиях подразумевается только общая площадь, и в open_plan. Попробуем более явно отделить такой вид пропусков (если он есть). Для кухни можно проставить 0, для жилой поставим общую
* В любом случае, остальные значения можно заменить расчётными (total_area мы знаем все)

```python
df.loc[df['kitchen_area'].isnull() & df['studio'] == True ] = df.loc[df['kitchen_area'].isnull() & df['studio'] == True ].fillna(0)
df.loc[df['living_area'].isnull() & df['studio'] == True ] = df.loc[df['living_area'].isnull() & df['studio'] == True ].fillna(df['total_area'])
df.info() # ставка не сыграла, но и данным мы особо не повредили. Используем план B
```

```python
# считаем медианные коэффициенты от общей площади для разных видов площадей
k1 = ( df[df['living_area'].notna()]['living_area'] / df[df['total_area'].notna()]['total_area']).dropna().median() 
k2 = ( df[df['kitchen_area'].notna()]['kitchen_area'] / df[df['total_area'].notna()]['total_area']).dropna().median()

# заменяем пропущенные значения
df['living_area'] = df['living_area'].fillna(df['total_area']*k1)
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area']*k2)
```

Вроде получилось. Дело за малым.

Следующий кандидат - days_exposition. В строках, где есть оба значения - конечная дата разница, значит дни считаются, когда объект продан. По умолчанию будем считать, что NaN - непроданная недвижимость.

По остальным переменным - особой закономерности их отсутствия выявить не удалось, оставим как есть. Вероятно это не повредит

## Расчёты и добавление результатов в таблицу

### Добавление столбцов даты

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
df.head(10)
```

### Категоризация этажа квартиры

```python
def floor_category(seq): # напишем функицю, которая сделает всё за меня
    if seq.loc['floor'] == 1:
        return 'первый'  # первый этаж
    elif seq.loc['floor'] == seq.loc['floors_total']:
        return 'последний' # последний этаж
    return 'другой' # другой этаж
   
df['floor_category'] = df.apply(floor_category, axis=1) # добавляем столбец   
df['floor_category'].value_counts()
```

#### Населённые пункты с самым большим числом объявлений
